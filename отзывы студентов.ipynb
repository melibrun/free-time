```python
def do_norm_pie(x):
    sp = list(x)
    result = {i: sp.count(i) for i in sp}
    result = dict(sorted(result.items()))
    color_s = {1:'tab:gray', 2:'tab:red', 3:'tab:purple', 4:'tab:olive', 5:'tab:green'}
    colors = []
    for ci in result.keys():
        for ki, vi in color_s.items():
            if ci == ki:
                colors.append(vi)

    u = str(x).split("/")[-1]
    plt.figure(figsize=(6, 6))
    plt.title(u, fontsize=20)
    
    plt.pie(result.values(), labels = result.keys(), colors = colors, textprops={'fontsize': 20})
    return colors
```

```python
for elem in list(marks.columns)[2::]:
    do_norm_pie(marks[elem])
```
